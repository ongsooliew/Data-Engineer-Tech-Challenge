```python
import requests

url = "https://api.covid19api.com/live/country/singapore"

payload={}
headers = {}

response = requests.request("GET", url, headers=headers, data=payload)

print(response.text)
```

```python
data=response.json()

data[0]
```

```python
import pandas as pd

df = pd.json_normalize(data)
df.head()
```

```python
df['Date']=pd.to_datetime(df['Date']).apply(lambda x: x.date())
df['Daily Confirmed Cases']= df['Confirmed'].diff()
```

```python
output= df[['Date', 'Daily Confirmed Cases' ]]
output.to_csv(r'data.csv')
```
